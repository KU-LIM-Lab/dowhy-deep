```python
# scripts/clean_graphs_1_10.py
import os
import pygraphviz as pgv

# graph_1.dot ~ graph_10.dot이 있는 디렉터리 경로
# 필요하면 절대경로로 수정하세요.
BASE_DIR = os.path(".")

# DAG에서 제거할 변수들
TO_REMOVE = {"CRQF", "JO_DNRG_RATE", "HOPE_JSCD1_NAME", "JHNT_CNT", "JHNT_MBN"}


def clean_graph(idx: int):
    fname = f"graph_{idx}.dot"
    path = os.path.join(BASE_DIR, fname)
    if not os.path.exists(path):
        print(f"[graph_{idx}] 파일 없음: {path}")
        return

    print(f"\n=== {fname} 정리 시작 ===")
    G = pgv.AGraph(path)

    removed_nodes = []
    # 1) 실제 변수 노드 제거
    for node in list(G.nodes()):
        name = node.get_name()
        if name in TO_REMOVE:
            removed_nodes.append(name)
            G.remove_node(node)

    removed_meta = []
    # 2) treatment_meta 노드 중 treatment_var가 제거 대상이면 같이 제거
    for node in list(G.nodes()):
        tv = node.attr.get("treatment_var")
        if tv in TO_REMOVE:
            removed_meta.append((node.get_name(), tv))
            G.remove_node(node)

    # 수정된 그래프를 덮어쓰기
    G.write(path)

    # 시각화 PNG 생성
    png_path = os.path.join(BASE_DIR, fname.replace(".dot", ".png"))
    G.layout(prog="dot")
    G.draw(png_path)

    print(f"  제거된 변수 노드: {removed_nodes or '없음'}")
    print(f"  제거된 treatment_meta 노드: {removed_meta or '없음'}")
    print(f"  시각화 저장: {png_path}")


def main():
    print(f"BASE_DIR = {BASE_DIR}")
    for i in range(1, 11):
        clean_graph(i)


if __name__ == "__main__":
    main()
```
